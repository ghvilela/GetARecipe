```python
from recipe_dict import recipe_dict
from main import quicksort, check_match


ingredients_list = []
for recipe in recipe_dict:
    for ingredient in eval(recipe["ingredients"]):
        if ingredient not in ingredients_list:
            ingredients_list.append(ingredient)
sorted_list = quicksort(ingredients_list)
            
print(len(sorted_list))
```

```python

clean_list = []
for i in sorted_list:
    new_i = str(i)
    clean_list.append(new_i)
    
print(clean_list)
```

```python
def check_match(input, text):
    if len(input) > len(text):
        return False
    match_count = 0
    for char in range(len(input)):
        if input[char] == text[char]:
            match_count += 1
        else:
            break
    if match_count == len(input):
        return True
    return False

repetition_list = {}
matched = []
for item in clean_list:
    if item not in matched:
        sample = item[:4]
        match_list = []
        for other in clean_list[1:]:
            if check_match(sample, other):
                matched.append(other)
                match_list.append(other)
        repetition_list[item] = match_list
            

print(repetition_list)
```
